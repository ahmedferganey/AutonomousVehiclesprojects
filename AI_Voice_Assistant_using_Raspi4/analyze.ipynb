```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data based on the provided content
data = {
    "Target Date": [
        "30-12-2023", "30-1", "30-2", "30-3", "30-4", "30-5", "30-6", "30-7", "30-8", "30-9", "30-10", "30-11", "30-12-2024", "30-1", "30-2"
    ],
    "Value added cost": [500, 500, 500, 500, 500, 600, 0, 0, 0, 0, 2250, 0, 1000, 0, 5300],
    "Personal": [320, 320, 320, 320, 320, 500, 0, 115, 2000, 130, 160, 270, 100, 100, 1230],
    "Mariam :D": [0, 0, 0, 0, 0, 0, 1000, 0, 0, 0, 0, 0, 0, 0, 0],
    "Transportation": [664, 664, 664, 664, 664, 750, 735, 800, 535, 140, 155, 0, 0, 0, 400],
    "In housing": [0, 0, 0, 0, 0, 500, 500, 135, 725, 550, 400, 1070, 230, 400, 240],
    "Occasions": [0, 0, 0, 0, 0, 950, 325, 0, 0, 0, 500, 0, 0, 0, 4200],
    "Breakfast": [0, 0, 0, 0, 200, 620, 205, 125, 150, 95, 250, 0, 190, 100, 200],
    "Oper": [0, 0, 0, 0, 200, 120, 30, 70, 0, 0, 0, 0, 0, 0, 0],
    "Losses": [816, 816, 816, 816, 816, 580, 265, 680, 475, 510, 600, 400, 520, 420, 500],
    "Weekend": [0, 0, 0, 0, 0, 0, 215, 220, 205, 275, 215, 180, 70, 230, 250]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate totals for each category
total_value_added = df["Value added cost"].sum()
total_personal = df["Personal"].sum()
total_mariam = df["Mariam :D"].sum()
total_transportation = df["Transportation"].sum()
total_in_housing = df["In housing"].sum()
total_occasions = df["Occasions"].sum()
total_breakfast = df["Breakfast"].sum()
total_oper = df["Oper"].sum()
total_losses = df["Losses"].sum()
total_weekend = df["Weekend"].sum()

# Group essential and non-essential costs
total_essential = total_personal + total_mariam + total_transportation 
total_non_essential = total_breakfast + total_oper + total_losses + total_weekend + total_in_housing + total_occasions

total_expense = total_value_added + total_essential + total_non_essential  # Calculate total expenses

# Create a summary DataFrame
summary_data = {
    "Category": ["Value Added Cost", "Non Value Added Cost 'Essential'", "Non Value Added Cost 'Non-Essential'"],
    "Total": [total_value_added, total_essential, total_non_essential]
}
summary_df = pd.DataFrame(summary_data)

# Calculate percentages
summary_df["Percentage"] = (summary_df["Total"] / total_expense) * 100

# Plotting
sns.set_theme(style="whitegrid")  # Modern theme
colors = ["#3498db", "#2ecc71", "#e74c3c"]  # Blue, Green, Red
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Category", y="Total", data=summary_df, palette=colors, edgecolor='black')

# Add value labels and percentage labels
for bar, percentage in zip(ax.patches, summary_df["Percentage"]):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 500, f"{round(yval, 2)}\n({round(percentage, 1)}%)", 
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Add title and labels
plt.title("Expense Breakdown", fontsize=16, fontweight='bold')
plt.ylabel("Total Expenses ($)", fontsize=12, fontweight='bold')
plt.xlabel("Category", fontsize=12, fontweight='bold')

# Add total expense label
plt.figtext(0.9, 0.2, f"Total: ${total_expense}", fontsize=14, fontweight='bold', color='black', 
            bbox=dict(facecolor='lightgray', alpha=0.5, edgecolor='black'))

plt.tight_layout()
plt.show()

# Print the summary table
print("Expense Summary:")
print(summary_df)
```
